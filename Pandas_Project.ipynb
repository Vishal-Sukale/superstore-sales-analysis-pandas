### Pandas Project - Superstore Sales Analysis

#### Problem Statement 
- Retail stores generate a large amount of sales data every day. It is difficult to understand sales performance, profit trends, and customer behavior without proper analysis. The problem of this project is to analyze the Superstore sales dataset using Pandas to understand sales patterns, regional performance, and product categories.


#### Objectives
The objectives of this project are:

- To analyze the Superstore dataset using Pandas.

- To understand sales and profit performance.

- To identify the best performing regions and product categories.

- To generate useful insights from the data.

#### Dataset Description
- The dataset used in this project is the Superstore Sales Dataset, which contains information about retail sales transactions. Each row represents a single order made by a customer.

- The dataset includes columns such as Order_ID, Order_Date, Segment, Region, Category, Sub_Category, Sales, Quantity, Discount, and Profit. These columns help analyze sales performance, customer segments, and product trends.

#### 1. Import Libraries
First, we import the necessary Python libraries.

```python
import pandas as pd
import numpy as np
```

#### 2. Load the Dataset


```python
df = pd.read_csv("superstore_dataset.csv")
```

- The dataset is loaded using the read_csv() function.

##### Preview of the Dataset

```python
df.head()
```

- It is used to display the first 5 rows of the dataset.

- It helps to quickly view and understand the structure of the data.

#### 3. Dataset Information
We check the dataset structure using:

```python
df.info()
```

- It is used to get a summary of the dataset, including column names, data types, and non-null values.

- It helps to understand the structure of the data and check missing values.

 

##### Shape of the Dataset

```python
df.shape
```

- This means the dataset contains 1000 rows and 10 columns.

#### 4. Statistical Summary
To understand numerical columns we use:

```python
df.describe()
```

- It provides statistical summary of numerical columns such as mean, minimum, maximum, and standard deviation.

- It helps to understand the distribution and overall behavior of the data.

#### 5. Checking Missing Values


```python
df.isnull().sum()
```

- It is used to check the total number of missing values in each column.

- It helps to identify which columns contain null values.

##### Fill Sales using Mean

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
```

- It fills missing values in the Sales column using the mean value.

- It is used for numerical data to replace missing values with the average.

##### Fill Quantity using Median

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
```

- It fills missing values in the Quantity column using the median value.

- Median is useful when the data may contain outliers.

##### Fill Segment using Mode

```python
df["Segment"] = df["Segment"].fillna(df["Segment"].mode()[0])
```

- It fills missing values in the Segment column using the mode (most frequent value).

- It is mainly used for categorical data.




```python
df.isnull().sum()
```

- It confirms that all missing values have been successfully filled in the dataset.

- The dataset is now clean and ready for further analysis or processing.

#### 6. Feature Engineering
Feature engineering helps create new useful columns from existing data.


##### Create Revenue Column

```python
df["Revenue"] = df["Sales"] * df["Quantity"]
```

- The Revenue column represents the total income generated from each order.

- It is calculated by multiplying Sales × Quantity.

#### 7. Category Analysis
To find the number of orders per category.

```python
df["Category"].value_counts()
```

- This shows the most frequently sold product category.

#### 8. Region Wise Sales Analysis
Using groupby() we analyze sales by region.

```python
df.groupby("Region")["Sales"].sum()
```

- This shows which region generates the highest revenue.

#### 9. Category Wise Performance
We analyze multiple metrics using agg().

```python
df.groupby("Category").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Quantity":"sum"
})
```

- It groups the data by Category and calculates the total Sales, Profit, and Quantity for each category.

- It helps to analyze the overall performance of each product category.

#### 10. Segment Wise Profit
Customer segments are analyzed using


```python
df.groupby("Segment")["Profit"].sum()
```

- This helps identify the most profitable customer group.

#### 11. Pivot Table Analysis
Pivot tables provide a multi-dimensional summary of the dataset.

```python
pd.pivot_table(
    df,
    values="Sales",
    index="Category",
    columns="Region",
    aggfunc="sum"
)
```

- It creates a pivot table showing total Sales for each Category across different Regions.

- It helps to compare sales performance of categories in different regions.

#### 12. Discount Impact Analysis
Discount can significantly affect profit.

```python
df.groupby("Discount")["Profit"].mean()
```

- It groups the data by Discount and calculates the average Profit for each discount level.

- It helps to understand how different discount levels affect profit.

#### 13. Correlation Analysis
Correlation shows relationships between variables.

```python
df[["Sales","Quantity","Discount","Profit"]].corr()
```

- It calculates the correlation between numerical columns such as Sales, Quantity, Discount, and Profit.

- It helps to understand the relationship between these variables (positive or negative relationship).

#### 14. Top 10 Profitable Orders


```python
df.sort_values(by="Profit", ascending=False).head(10)
```

- It sorts the dataset by Profit in descending order.

- It displays the top 10 orders with the highest profit.

#### 15. Export Clean Dataset


```python
df.to_csv("clean_superstore_data.csv", index=False)
```

- It is used to save the cleaned or processed dataset into a CSV file.

- index=False ensures that the row index is not saved in the CSV file.
